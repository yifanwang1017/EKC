```python
import pandas as pd
```

```python
data = pd.read_excel('推断数据.xlsx')
```

```python
data.head()
```

```python
data1 = data[data['crash']==1]
```

```python
print(len(data1))
```

```python
data1 = data1.sort_values(by='p',ascending=0)
data1.head(10)
```

```python
t = data1.iloc[0:100]
print(len(t))
m = t.groupby(['time of week']).count().reset_index()
print(m)
```

```python
data2 = data[data['crash']==0]
print(len(data2))
```

```python
s = pd.merge(data1,data2,on=['volume','temperature','season','weather','daytime','time of week'])
print(len(s))
print(s.columns)
s.head()
```

```python
s['crash_p'] = s['p_x']/(s['p_x']+s['p_y'])
s = s.sort_values(by=['p_x','crash_p'],ascending=False)
s.head(20)
```
